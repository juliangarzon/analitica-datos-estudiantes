```python
from pathlib import Path

import pandas as pd
import plotly.express as px

DATA_PATH = Path('..') / 'data' / 'calidad_aire_risaralda.csv'

df = pd.read_csv(DATA_PATH)
df = df.rename(
    columns={
        'Municipio': 'municipio',
        'Estacion': 'estacion',
        'Fecha': 'fecha',
        'Diametro aerodinamico': 'diametro',
        'Medicion': 'medicion',
    }
)

df['fecha'] = pd.to_datetime(df['fecha'])
df['medicion'] = pd.to_numeric(df['medicion'], errors='coerce')
df = df.dropna(subset=['fecha', 'medicion']).sort_values('fecha')

df.head()
```

## Task 1. Simula los filtros del sidebar

Selecciona un subconjunto de municipios, un subconjunto de partículas y un rango de fechas.


```python
selected_municipios = ['Pereira', 'Dosquebradas']
selected_particles = ['PM10', 'PM2.5']
start_date = pd.Timestamp('2019-01-01')
end_date = pd.Timestamp('2023-07-30')

filtered_df = df[
    df['municipio'].isin(selected_municipios)
    & df['diametro'].isin(selected_particles)
    & df['fecha'].between(start_date, end_date)
].copy()

filtered_df.head()
```
