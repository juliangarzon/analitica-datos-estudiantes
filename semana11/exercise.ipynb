**Dataset:** `../data/calidad_aire_risaralda.csv`


## Parte 1. Carga y entiende el dataset

Antes de graficar, identifica:

1. qué representa cada fila
2. qué filtros serían útiles
3. qué métrica numérica será el centro del dashboard


```python
from pathlib import Path

import pandas as pd
import plotly.express as px

DATA_PATH = Path('data') / 'calidad_aire_risaralda.csv'

df = pd.read_csv(DATA_PATH)
df = df.rename(
    columns={
        'Municipio': 'municipio',
        'Estacion': 'estacion',
        'Fecha': 'fecha',
        'Diametro aerodinamico': 'diametro',
        'Medicion': 'medicion',
    }
)

df['fecha'] = pd.to_datetime(df['fecha'])
df['medicion'] = pd.to_numeric(df['medicion'], errors='coerce')
df = df.dropna(subset=['fecha', 'medicion']).sort_values('fecha')

df.head()
```

```python
df.describe(include='all')
```

## Parte 2. Línea temporal

Crea un gráfico de línea con el promedio de medición por fecha y color por tipo de partícula.


```python
trend_df = (
    df.groupby(['fecha', 'diametro'], as_index=False)['medicion']
    .mean()
    .sort_values('fecha')
)

fig = px.line(
    trend_df,
    x='fecha',
    y='medicion',
    color='diametro',
    markers=True,
    title='Average particulate matter by day'
)
fig.show()
```

## Parte 3. Comparación entre municipios

Crea un gráfico de barras con el promedio de medición por municipio.


```python
municipio_df = (
    df.groupby('municipio', as_index=False)['medicion']
    .mean()
    .sort_values('medicion', ascending=False)
)

fig = px.bar(
    municipio_df,
    x='municipio',
    y='medicion',
    color='municipio',
    title='Average particulate matter by municipality'
)
fig.show()
```

## Parte 4. Distribución por estación

Crea un box plot que muestre cómo cambia la medición según estación y tipo de partícula.


```python
fig = px.box(
    df,
    x='estacion',
    y='medicion',
    color='diametro',
    title='Distribution of measurements by station'
)
fig.show()
```
